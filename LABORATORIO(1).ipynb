Libraries

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

### Reading the csv

```python
df = pd.read_csv('dol.csv')
#df = df.iloc[:5000]
```

### Creating columns

```python
df['return'] = df['close'].pct_change() * 100
df['inclination'] = df['return'].rolling(window=4).apply(lambda x: np.polyfit(range(4), x, 1)[0], raw=True)
df['return_last_min'] = df['return'].shift(1)
df['return_candle'] = (df['open'] / df['close']) - 1
df['return_next_min'] = df['return'].shift(-1)
df = df.iloc[10:]



x = df[['inclination', 'return_last_min', 'close']].iloc[:1000]
y = df['return_next_min'].iloc[:1000]
```

### Linear Regression

```python
df = pd.read_csv('dol_return.csv')


# Setting features(y) and target(x)
x = df[['inclination', 'return_last_min', 'close']].iloc[:1000]
y = df['return_next_min'].iloc[:1000]


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and train the linear regression model
model = LinearRegression()
model.fit(x_train, y_train)

# Making predictions
predictions = model.predict(x_test)

# Calculating mean squared error
mse = mean_squared_error(y_test, predictions)
print("Mean Squared Error:", mse)

# Making predictions for the whole data
df['predicted_return_next_min'] = model.predict(df[['inclination', 'return_last_min', 'close']])
```
